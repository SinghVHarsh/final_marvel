```python
import hashlib
import requests
import datetime


import pandas as pd

timestamp = datetime.datetime.now().strftime('%Y-%m-%d%H:%M:%S')
pub_key = input('enter the public key')
priv_key = input('enter the private key')


name=input()
def hash_params():


    """ Marvel API requires server side API calls to include
    md5 hash of timestamp + public key + private key """

    hash_md5 = hashlib.md5()
    hash_md5.update(f'{timestamp}{priv_key}{pub_key}'.encode('utf-8'))
    hashed_params = hash_md5.hexdigest()

    return hashed_params
df_new=pd.DataFrame()
for i in range(0,4):

    params = {'ts': timestamp, 'apikey': pub_key, 'hash': hash_params(),'nameStartsWith':name,'limit':100,'offset': 100*i}
    res = requests.get('https://gateway.marvel.com:443/v1/public/characters',params=params)


    call = res.json()
    data=(call['data']['results'])
    df1=pd.json_normalize(data)
    df_new=df_new.append(df1,ignore_index=True)
df_new[['name','events.available','series.returned','stories.available','comics.available','id']]
   



```
